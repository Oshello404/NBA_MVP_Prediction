```python
# Find average precision (AP), a common scoring metric
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5) # Top 5 MVP ranks, k = 5
    predicted = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index,row in predicted.iterrows(): # Go down the rows
        if row["Player"] in actual["Player"].values: # If Predicted rank in top 5 of Actual rank,
            found += 1 # Found +1
            ps.append(found / seen) # Division to find the precision value
        seen += 1

    return sum(ps) / len(ps) # Average Precision (AP)
```

```python
def add_ranks(predictions):
    predictions = predictions.sort_values("predictions", ascending=False)
    predictions["Predicted_Rk"] = list(range(1,predictions.shape[0]+1))
    predictions = predictions.sort_values("Share", ascending=False)
    predictions["Rk"] = list(range(1,predictions.shape[0]+1))
    predictions["Diff"] = (predictions["Rk"] - predictions["Predicted_Rk"])
    return predictions
```

```python
# Backtest to use all historical data before a certain year, then repeat for next year
def backtest(stats, model, years, predictors):
    aps = []
    all_predictions = []
    for year in years:
        train = stats[stats["Year"] < year] # Training data is all historical data before that year
        test = stats[stats["Year"] == year] # Test data is just that year
        model.fit(train[predictors],train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_predictions)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
def backtest(stats, model, years, predictors):
    aps = []
    all_predictions = []
    mvp_acc = []
    start_year = years[0]

    for year in years:
        train = stats[stats["Year"] < year].copy()
        test = stats[stats["Year"] == year].copy()

        sc.fit(train[predictors])
        train[predictors] = sc.transform(train[predictors])
        test[predictors] = sc.transform(test[predictors])

        model.fit(train[predictors],train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)

        combination = add_ranks(combination)
        if combination.iloc[0]["Predicted_Rk"] == 1:
          print(year, " MVP is ", combination.iloc[0]['Player'])
          mvp_acc.append(1)
        else:
          print(year,  " MVP is NOT ", combination.loc[combination['Predicted_Rk'] == 1, 'Player'])
          mvp_acc.append(0)

        all_predictions.append(combination)
        aps.append(find_ap(combination))

    return sum(aps) / len(aps), aps, pd.concat(all_predictions), start_year, sum(mvp_acc)/len(mvp_acc)
```

```python
def mvp_rk(all_predictions):
    return all_predictions[all_predictions['Rk'] == 1]['Predicted_Rk'].tolist()
```
